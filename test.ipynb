```python
import numpy as np
import pandas as pd
import duckdb
```

```python
df = pd.DataFrame()
ids = np.array(range(1,101))
data = np.random.rand(100)
data = np.multiply(data, 10)
df["id"] = ids
df["rating"] = data
df.info()
```

```python
duckdb.sql("CREATE TABLE database AS SELECT * FROM df;")
```

```python
result = duckdb.query("SELECT * FROM database").df()
result
```

```python
duckdb.query("INSERT INTO database VALUES (101, 6.9)")
```
